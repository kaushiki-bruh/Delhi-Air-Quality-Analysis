```python
import pandas as pd

df = pd.read_csv("final_dataset.csv")

df.head()
df.info()
df = df.drop_duplicates()
```

```python
plt.figure()
plt.plot(df["date_final"], df["AQI"], color='magenta')
plt.title("AQI Over Time")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.show()
```

```python
import matplotlib.pyplot as plt
df.groupby('Month')['AQI'].mean().plot(kind='bar')
plt.title('Average AQI by Month')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("final_dataset.csv")
df["date_final"] = pd.to_datetime(df["date_final"])

plt.figure()

plt.plot(df["date_final"], df["PM2.5"], label="PM2.5")
plt.plot(df["date_final"], df["PM10"], label="PM10")
plt.plot(df["date_final"], df["NO2"], label="NO2")
plt.plot(df["date_final"], df["SO2"], label="SO2")
plt.plot(df["date_final"], df["CO"], label="CO")

plt.title("Pollutant Concentration Over Time")
plt.xlabel("Date")
plt.ylabel("Concentration")
plt.legend()
plt.show()
```

```python
clean_days = df.nsmallest(10, "AQI")

plt.figure()
plt.bar(clean_days["date_final"].astype(str), clean_days["AQI"], color='purple')
plt.title("Top 10 Lowest AQI Days")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
plt.hist(df["AQI"], color='orange')
plt.title("AQI Distribution")
plt.xlabel("AQI")
plt.ylabel("Frequency")
plt.show()
```
